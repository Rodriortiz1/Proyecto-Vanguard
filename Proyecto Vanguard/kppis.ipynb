## TASA DE ERROR POR STEP

```python
# Asegurar que la columna date_time sea del tipo datetime
df_control['date_time'] = pd.to_datetime(df_control['date_time'])

# Crear un mapeo de pasos a nÃºmeros para facilitar la comparaciÃ³n
step_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}
df_control['step_num'] = df_control['process_step'].map(step_map)

# Ordenar el DataFrame para garantizar el correcto cÃ¡lculo de pasos previos
df_control = df_control.sort_values(by=['client_id', 'date_time'])

# Identificar retrocesos, excluyendo el primer paso para cada cliente
df_control['prev_step_num'] = df_control.groupby('client_id')['step_num'].shift(1)
df_control['error'] = (df_control['prev_step_num'] > df_control['step_num']) & (df_control['process_step'] != 'start')

# Contar errores por paso, excluyendo 'start'
errors_per_step = df_control[df_control['error'] & (df_control['process_step'] != 'start')].groupby('process_step').size()
total_per_step = df_control[df_control['process_step'] != 'start'].groupby('process_step').size()

# Calcular la tasa de error por paso
error_rate_per_step = (errors_per_step / total_per_step).fillna(0)

print("Errores por paso:")
print(errors_per_step)
print("\nTotal por paso:")
print(total_per_step)
print("\nTasa de error por paso:")
print(error_rate_per_step)
```

## DURACION TOTAL PROMEDIO DE CADA STEP

```python
# Asegurar que la columna date_time sea del tipo datetime
df_test['date_time'] = pd.to_datetime(df_test['date_time'])

# Ordenar el DataFrame por client_id, visitor_id, visit_id y date_time
df_test = df_test.sort_values(by=['client_id', 'visitor_id', 'visit_id', 'date_time'])

# Calcular la diferencia de tiempo en segundos entre cada fila dentro de cada client_id, visitor_id y visit_id
df_test['time_diff'] = df_test.groupby(['client_id', 'visitor_id', 'visit_id'])['date_time'].diff().dt.total_seconds()

# Llenar los valores NaN con 0 (para el primer paso de cada cliente)
df_test['time_diff'] = df_test['time_diff'].fillna(0)

# Agrupar por client_id y process_step, y sumar las diferencias de tiempo
time_per_step_test = df_test.groupby(['client_id', 'process_step'])['time_diff'].sum().reset_index()

# Pivotear el DataFrame para tener process_step como columnas
time_per_step_pivot_test = time_per_step_test.pivot(index='client_id', columns='process_step', values='time_diff').fillna(0).reset_index()

# Calcular la duraciÃ³n promedio por paso
mean_duration_per_step = time_per_step_pivot_test.mean().drop('client_id')

# Convertir la duraciÃ³n promedio de segundos a minutos
mean_duration_per_step_test_minutes = mean_duration_per_step / 60

# Mostrar las duraciones promedio por paso en minutos
print("DuraciÃ³n promedio por paso para el grupo test (en minutos):")
print(mean_duration_per_step_test_minutes)
```

```python
import matplotlib.pyplot as plt

# VisualizaciÃ³n de resultados
plt.figure(figsize=(10, 6))
mean_duration_per_step_test_minutes.plot(kind='bar')
plt.xlabel('Process Step')
plt.ylabel('Average Duration (minutes)')
plt.title('Average Duration per Process Step in test group (in minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as st 

#H0 = tasa de finalizacion del grupo test = tasa de finalizacion del grupo control 
#H1 = tasa de finaliazcion del grupo test != tasa de finalizacion del grupo control 

stat, p_value = st.ttest_ind (df_control["Completion rate"],df_test["Completion rate"], equal_var = False)
print(stat)
print(p_value)


# DeterminaciÃ³n
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipÃ³tesis nula. Hay una diferencia estadÃ­sticamente significativa entre las tasas de finalizaciÃ³n.")
else:
    print("No podemos rechazar la hipÃ³tesis nula. No hay suficiente evidencia para afirmar que hay una diferencia significativa entre las tasas de finalizaciÃ³n.")
```

```python
# Calcular las tasas de finalizaciÃ³n promedio
mean_control = df_control['Completion rate'].mean()
mean_test = df_test['Completion rate'].mean()

# Definir el umbral del 5%
threshold = 0.05
mean_control_with_threshold = mean_control + threshold

# Realizar la prueba t de una cola (one-tailed test)
stat, p_value = st.ttest_ind(df_test['Completion rate'], df_control['Completion rate'], equal_var=False)

#La tasa de finalizaciÃ³n del grupo test no es al menos un 5% mayor que la tasa de finalizaciÃ³n del grupo control.
#La tasa de finalizaciÃ³n del grupo test es al menos un 5% mayor que la tasa de finalizaciÃ³n del grupo control.


# Ajustar el p-value para una prueba de una cola
p_value_one_tailed = p_value / 2

# Resultados
print(f"EstadÃ­stico t: {stat}")
print(f"Valor p (una cola): {p_value_one_tailed}")

# DeterminaciÃ³n
alpha = 0.05
if mean_test > mean_control_with_threshold and p_value_one_tailed < alpha:
    print("Rechazamos la hipÃ³tesis nula. El aumento en la tasa de finalizaciÃ³n es al menos un 5% mayor y es estadÃ­sticamente significativo.")
else:
    print("No podemos rechazar la hipÃ³tesis nula. El aumento en la tasa de finalizaciÃ³n no es al menos un 5% mayor o no es estadÃ­sticamente significativo.")
```

```python
import scipy.stats as stats

contingency_table = pd.crosstab(df_merged_1['gender'], df_merged_1['Variation'])

print("Tabla de Contingencia:")
print(contingency_table)

# Realizar la prueba de chi-cuadrado de independencia
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nEstadÃ­stico chi-cuadrado: {chi2}")
print(f"Valor p: {p}")
print(f"Grados de libertad: {dof}")
print("Tabla esperada:")
print(expected)

# HipÃ³tesis nula (H0): No hay asociaciÃ³n entre el gÃ©nero y el tipo de proceso (nuevo o antiguo).
# HipÃ³tesis alternativa (H1): Hay una asociaciÃ³n significativa entre el gÃ©nero y el tipo de proceso.

# DeterminaciÃ³n
alpha = 0.05
if p < alpha:
    print("Rechazamos la hipÃ³tesis nula. Hay una asociaciÃ³n significativa entre el gÃ©nero y el tipo de proceso.")
else:
    print("No podemos rechazar la hipÃ³tesis nula. No hay suficiente evidencia para afirmar que hay una asociaciÃ³n significativa entre el gÃ©nero y el tipo de proceso.")
```
