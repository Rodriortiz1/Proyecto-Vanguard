## Analisis de la columna Age

```python
mean_age = df2['clnt_age'].mean().round(2)
min_age = df2['clnt_age'].min()
max_age = df2['clnt_age'].max()
range_age = max_age - min_age

mean_age, min_age,max_age,range_age
```

La media de edad es 46.44 

El minimo de edad es de 13,5 

El maximo es de 96

Lo que da un rango de edad de 82,5 

```python
df2.clnt_age.unique()
```

```python
def categorize_age(age):
    if 13 <= age < 30:
        return 'young'
    elif 30 <= age < 60:
        return 'adult'
    elif 60 <= age < 100:
        return 'senior'
    else:
        return 'fuera de rango'

# Aplicar la funciÃ³n de mapeo a la columna 'edad' para crear la nueva columna 'rango de edad'
df2['grupo_edad'] = df2['clnt_age'].map(categorize_age)

# Mostrar el DataFrame resultante
df2.head(10)
```

```python
my_table = pd.crosstab(index = df2["grupo_edad"],  
                              columns="count")      
my_table
```

```python
import seaborn as sns
sns.barplot(x=my_table.index, y=my_table["count"], palette="Set3");
```

Podemos ver que el grupo donde la mayoria de usuarios se encuentra es adulto

## Analisis antiguedad por aÃ±os

```python
mean_yr = df2['clnt_tenure_yr'].mean().round(2)
min_yr = df2['clnt_tenure_yr'].min()
max_yr = df2['clnt_tenure_yr'].max()
range_yr = max_yr - min_yr

mean_yr, min_yr, max_yr, range_yr
```

La media de antiguedad es 12 aÃ±os

el min es 2 aÃ±os 

el max es 62 aÃ±os 

Rango de antiguedad de 60 aÃ±os  


```python
df2.clnt_tenure_yr.unique()
```

```python
def categorize_age(tenure):
    if 0 <= tenure < 10:
        return 'nuevo'
    elif 10 <= tenure < 30:
        return 'regular'
    elif 30 <= tenure < 65:
        return 'fiel'
    else:
        return 'fuera de rango'

# Aplicar la funciÃ³n de mapeo a la columna 'edad' para crear la nueva columna 'rango de edad'
df2['grupo_antiguedad'] = df2['clnt_tenure_yr'].map(categorize_age)

# Mostrar el DataFrame resultante
df2.head(10)
```

```python
my_table = pd.crosstab(index = df2["grupo_antiguedad"],  
                              columns="count")      
my_table
```

```python
import seaborn as sns
sns.barplot(x=my_table.index, y=my_table["count"], palette="Set3");
```

Podemos ver que la mayoria de usuarios son regulares (de 10 a 30 aÃ±os) sin embargo hay una gran cantidad de clientes nuevos (de 0 a 10 aÃ±os)
